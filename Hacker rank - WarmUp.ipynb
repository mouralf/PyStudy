```python
def hourglassSum(arr):
    maxHour = 0

    if (verifHours(arr)):

        for i in range(len(arr)):
            #print(arr[i])
            for j in range(len(arr)):
                #print(arr[i][j])
                if(i+2 < len(arr) and j+2 < len(arr)):
                    
                    a = arr[i][j]
                    b = arr[i][j+1]
                    c = arr[i][j+2]
                    d = arr[i+1][j+1]
                    e = arr[i+2][j]
                    f = arr[i+2][j+1]
                    g = arr[i+2][j+2]
                    
                    print(str(a) + "\t " + str(b) + "\t " + str(c) + "\n \t" + str(d)+ "\n " + str(e) + "\t " + str(f) + "\t " + str(g))
                    
                    
                    sum = a+b+c+d+e+f+g
                    print("\n Sum:" + str(sum))
                    if (sum>maxHour):
                        maxHour = sum
    else:
        maxHour = None
    
    return maxHour



def verifHours(arr):
    const = True

    for i in range(len(arr)):
        for j in range(len(arr)):
            if not (arr[i][j]>=-9 and arr[i][j]<=9):
                const = False
                
            if not (i>=0 and i<=5):
                const = False
            if not (j>=0 and j<=5):
                const = False

    return const

array_2d = [
    [-9, -9, -9, 1, 1, 1],
    [0, -9, 0, 4, 3, 2],
    [-9, -9, -9, 1, 2, 3],
    [0, 0, 8, 6, 6, 0],
    [0, 0, 0, -2, 0, 0],
    [0, 0, 1, 2, 4, 0]
]

hourglassSum(array_2d)
```
